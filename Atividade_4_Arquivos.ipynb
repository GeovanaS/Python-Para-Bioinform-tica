```
arquivo = open('Arquivo_de_texto.txt','w')
for numero in range(1000):
    numero2 = numero * 2
    arquivo.write("O resultado da multiplicação de %s por 2 é: " %numero + str(numero2) + "\n")
arquivo.close()
```

```
arquivo = open('Arquivo_de_texto.txt','a')
arquivo.write("Nova linha\n")
arquivo.write("Segunda nova linha")
arquivo.close()
```
